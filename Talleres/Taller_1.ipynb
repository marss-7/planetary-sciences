```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calculate the mass of the protoplanet

years = np.linspace(0, 100, 10)  # Time in years
t = 3.154e7  # Time in seconds (1 year)
G = 6.67430e-8  # Gravitational constant in cm^3 g^-1 s^-2
M_sun = 1.989e33  # Mass of the Sun in grams
distance = 1.496e13  # Distance from the Sun to the protoplanet in cm
miu = 5e-3  # Viscosity parameter (dimensionless)
density_gas = 10**-9  # Gas density in g/cm^3


v_keplerian = np.sqrt(G * M_sun / distance)  # Keplerian velocity in cm/s
v_gas = v_keplerian * np.sqrt(1 - miu)  # Gas velocity in cm/s
relative_velocity = v_keplerian - v_gas  # Relative velocity in cm/s

def calculate_mass(year):
    mass = (density_gas * np.pi * relative_velocity * t * year)
    return mass

masses = calculate_mass(years)
masses, relative_velocity
```

```python
# plot mass * R^2 versus time
plt.figure(figsize=(10, 6))
plt.plot(years, masses, marker='o')
plt.title('Mass of the Protoplanet Over Time')
plt.xlabel('Time (years)')
plt.ylabel('Mass (g) * R^2')
plt.grid()
plt.show()
```

```python
# R^2

def calculate_radius(mass):
    radius = (3 * mass / (4 * np.pi * 3))
    return radius
radii = calculate_radius(masses)

# plot radius versus time
plt.figure(figsize=(10, 6))
plt.plot(years, radii, marker='o')
plt.title('Radius of the Protoplanet Over Time')
plt.xlabel('Time (years)')
plt.ylabel('Radius (cm)')
plt.grid()
plt.show()
```
