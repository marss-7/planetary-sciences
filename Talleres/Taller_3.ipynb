```python
# Rotational of the central bodies

I_sol = 0.07 * Sol_m * (696340e3)**2
I_jupiter = 0.25 * Jupiter_m * (69911e3)**2
I_saturn = 0.21 * Saturn_m * (58232e3)**2
I_uranus = 0.22 * Uranus_m * (25362e3)**2

# Definimos los periodos en segundos
T_sol = 25.38 * 24 * 3600
T_jupiter = 9.92 * 3600
T_saturn = 10.65 * 3600
T_uranus = 17.24 * 3600

# Calculamos las velocidades angulares (rad/s)
omega_sol = (2 * np.pi) / T_sol
omega_jupiter = (2 * np.pi) / T_jupiter
omega_saturn = (2 * np.pi) / T_saturn
omega_uranus = (2 * np.pi) / T_uranus

# Calculamos L_rotacional
L_sol = I_sol * omega_sol
L_jupiter = I_jupiter * omega_jupiter
L_saturn = I_saturn * omega_saturn
L_uranus = I_uranus * omega_uranus
```

```python
print("Angular momentum of the Sun: ", L_sol, "kg m^2/s")
print("Angular momentum of Jupiter: ", L_jupiter, "kg m^2/s")
print("Angular momentum of Saturn: ", L_saturn, "kg m^2/s")
print("Angular momentum of Uranus: ", L_uranus, "kg m^2/s")

print("Angular momentum of the planets in the Solar system: ", L_planets, "kg m^2/s")
print("Angular momentum of the 8 most massive bodies in the Jovian system: ", L_jovian, "kg m^2/s")
print("Angular momentum of the 8 most massive bodies in the Saturnian system: ", L_saturnian, "kg m^2/s")
print("Angular momentum of the 8 most massive bodies in the Uranian system: ", L_uranian, "kg m^2/s")

print(f"Solar system angular momentum ratio: {np.linalg.norm(L_planets) / np.linalg.norm(L_sol)}")
print(f"Jovian system angular momentum ratio: {np.linalg.norm(L_jovian) / np.linalg.norm(L_jupiter)}")
print(f"Saturnian system angular momentum ratio: {np.linalg.norm(L_saturnian) / np.linalg.norm(L_saturn)}")
print(f"Uranian system angular momentum ratio: {np.linalg.norm(L_uranian) / np.linalg.norm(L_uranus)}")
```
